## 1. Distance metric

What is distance metric? 

In machine learning, a distance metric, also known as a similarity metric or similarity function, is a mathematical function that defines the distance or similarity between two points in a multi-dimensional space. The distance metric is used to measure the similarity between objects, such as data points or feature vectors, and is often used in algorithms that perform clustering, classification, and anomaly detection.

There are several commonly used distance metrics in machine learning, including:

- **Euclidean distance**: This is the most common distance metric and is simply the straight-line distance between two points in a multi-dimensional space.

- **Manhattan distance**: Also known as the L1 norm, this distance metric calculates the absolute differences between the coordinates of two points and sums them up.

- **Cosine similarity**: This is a measure of similarity between two non-zero vectors of an inner product space.

- **Jaccard similarity**: This is a measure of the similarity between two sets, and is defined as the size of the intersection divided by the size of the union of the sets.

- **Mahalanobis distance**: This is a measure of the distance between two points in a multi-dimensional space, where the distance is adjusted for the correlations between the dimensions.

The choice of a distance metric depends on the nature of the data and the problem being solved, and often requires trial-and-error and careful consideration.

Given $x, y=$ two points in Euclidean $\mathrm{n}$-space

$\mathbf{x}= (x_1, x_2, ..., x_n)$ 

$\mathbf{y}= (y_1, y_2, ..., y_n)$

$n: \quad n$-space

##### 1.1 Formula for some distance metric

##### Euclidean distance:

$$
d(\mathbf{x}, \mathbf{y})=\sqrt{\sum_{i=1}^n\left(x_i-y_i\right)^2}
$$

##### Manhattan distance

$$
d(\mathbf{x}, \mathbf{y})={\sum_{i=1}^n | x_i - y_i |}
$$

##### Minkowski distance

$$
d(\mathbf{x}, \mathbf{y})=({\sum_{i=1}^n | x_i - y_i |^{p}})^{\frac{1}{p}}
$$

Now lets define a simple euclidead distance function

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2)**2))
```

##### 2.1 One-hot encoding for categorical variable

2.1.1 One-hot encoding is a way to represent a **categorical variable** as a **numerical vector**. 

In a one-hot encoded representation, each category is represented as a **binary vector with a length equal to the number of categories**. Each vector has a value of 1 in the index corresponding to the represented category, and 0s everywhere else. 

This allows machine learning algorithms to work with categorical data, as they typically expect numerical input.

2.1.2 For example, if we have a categorical variable with 3 categories, "red", "green", and "blue", we can represent these categories as the following one-hot encoded vectors:

    "red" : [1, 0, 0]
    "green": [0, 1, 0]
    "blue" : [0, 0, 1]

This encoding is often used in natural language processing and computer vision applications, where the inputs can be categorical in nature, such as words in a sentence or object classes in an image.

2.1.3 Example code:

```python
"""
This code defines a function one_hot_encode that takes a DataFrame and a column name as input, 
performs one-hot encoding on the specified column, and returns the modified DataFrame with the encoded column(s). 
The code uses the get_dummies function from the Pandas library to perform the encoding, 
and the join function to merge the original DataFrame with the encoded one-hot DataFrame.
"""
def one_hot_encode(data, column):
    one_hot = pd.get_dummies(data[column])
    data = data.drop(column, axis=1)
    data = data.join(one_hot)
    return data
```

##### 2.2 Normalization / Rescaling in the k-nearest neighbors for continuos features

2.2.1 In the k-nearest neighbors (KNN) algorithm, the distance between two points is used as a measure of similarity. This means: 

        The scale of the features can have a significant impact on the results of KNN.

2.2.2 For example, consider two features x1 and x2, where **x1 is measured in inches and x2 is measured in centimeters**. The Euclidean distance between two points (x1, x2) and (y1, y2) is given by the formula:

        distance = sqrt((x1 - y1)^2 + (x2 - y2)^2)

Since x2 is in centimeters, it will have a much larger scale than x1, which means that the contribution of x2 to the distance will be much larger than the contribution of x1. **This can result in the algorithm placing more weight on x2 than it does on x1, even if x1 is more relevant for the task at hand**.

2.2.3 Normalizing the data before using KNN helps to avoid this problem by transforming the features to have zero mean and unit variance. **This ensures that all features contribute equally to the distance calculation, and the scale of each feature does not have an undue influence on the results.**

2.2.4 Example code:

```python
"""

This code defines a function rescale that takes a Pandas DataFrame as input and returns the rescaled data. 
The code uses the min and max methods to calculate the minimum and maximum of each column, 
and subtracts the minimum from each column and divides by the range (i.e., the difference between the maximum and minimum). 
This rescales the data to have a range of 0 to 1 for each feature.

"""

def rescale(data):
    return (data - data.min()) / (data.max() - data.min())

data = pd.DataFrame({'A': [1, 4, 7], 'B': [2, 5, 8], 'C': [3, 6, 9]})
print(data)
rescaled_data = rescale(data)
print('after rescaling')
print(rescaled_data)
```

## 3, K-Nearest Neighbor

##### 3.1 KNN from scratch. 

The comparison may demonstrate the usefulness of using matrix operations in python as we introduced in Notebook 1

```python
from tqdm import tqdm
class KNN_v1:
    def __init__(self, k):
        self.k = k
        
    def euclidean_distance(self, point1, point2):
        return np.sqrt(np.sum((point1 - point2)**2))
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self, test_points):
        predictions = []
        for test_point in tqdm(test_points):
            distances = [self.euclidean_distance(test_point, x) for x in self.X_train]
            nearest_indices = np.argsort(distances)[:self.k]
            nearest_labels = [self.y_train[i] for i in nearest_indices]
            predictions.append(Counter(nearest_labels).most_common(1)[0][0])
        return predictions
```

```python
knn = KNN_v1(k = 10)
knn.fit(trainX, trainY)
predicted_labels = knn.predict(testX)
print('Accuracy:', np.mean(predicted_labels == testY))
```

```python
class KNN_v2:
    def __init__(self, k):
        self.k = k
        
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self, test_points):
        distances = np.sqrt(np.sum((self.X_train - test_points[:, np.newaxis, :])**2, axis=2))
        nearest_indices = np.argsort(distances, axis=1)[:, :self.k]
        nearest_labels = self.y_train[nearest_indices]
        predictions = [Counter(row).most_common(1)[0][0] for row in nearest_labels]
        return predictions
```

```python
knn = KNN_v2(k = 10)
knn.fit(trainX, trainY)
predicted_labels = knn.predict(testX)
print('Accuracy:', np.mean(predicted_labels == testY))
```

## 4, The bias-variance trade-off in KNN

##### 4.1 What is bias-variance trade-off? 

The bias-variance tradeoff is a fundamental concept in machine learning and applies to the k-nearest neighbor (KNN) algorithm as well.

- **Bias** refers to the error that is introduced by approximating a complex relationship between the features and the target variable. A high bias model will oversimplify the relationship and result in underfitting, meaning that the model will perform poorly on both the training data and unseen data.

- **Variance** refers to the error that is introduced by the model being too sensitive to the noise in the training data. A high variance model will overfit the data, meaning that it will perform well on the training data but poorly on unseen data.


\begin{aligned}
\mathrm{E}\left[\left(Y-\widehat{f}\left(x_0\right)\right)^2\right] & =\mathrm{E}\left[\left(Y-f\left(x_0\right)+f\left(x_0\right)-\mathrm{E}\left[\widehat{f}\left(x_0\right)\right]+\mathrm{E}\left[\widehat{f}\left(x_0\right)\right]-\widehat{f}\left(x_0\right)\right)^2\right] \\
& =\mathrm{E}\left[\left(Y-f\left(x_0\right)\right)^2\right]+\mathrm{E}\left[\left(f\left(x_0\right)-\mathrm{E}\left[\hat{f}\left(x_0\right)\right]\right)^2\right]+\mathrm{E}\left[\left(E\left[\widehat{f}\left(x_0\right)\right]-\widehat{f}\left(x_0\right)\right)^2\right]+\text { Cross Terms } \\
& =\underbrace{\mathrm{E}\left[\left(Y-f\left(x_0\right)\right)^2\right]}_{\text {Irreducible Error }}+\underbrace{\left(f\left(x_0\right)-\mathrm{E}\left[\widehat{f}\left(x_0\right)\right]\right)^2}_{\text {Bias }^2}+\underbrace{\mathrm{E}\left[\left(\widehat{f}\left(x_0\right)-\mathrm{E}\left[\widehat{f}\left(x_0\right)\right]\right)^2\right]}_{\text {Variance }}
\end{aligned}

##### 4.2 bias and variance in the KNN
- the **bias is relatively low** because the model can fit complex relationships between the features and the target variable. 

- the **variance is relatively high** because the model is very sensitive to the noise in the training data. 

For any given point, the prediction is based on the average of the target variable for the k-nearest neighbors, and **small fluctuations in the training data can have a big impact on the prediction**.


##### 4.3 bias-variance trade-off in the KNN

The tradeoff in k-NN can be controlled by the value of k. A small k value will result in a high variance and low bias model, while a large k value will result in a low variance and high bias model. The goal is to find the value of k that results in the best tradeoff between bias and variance for a given dataset.

##### 4.4 Example, bias, variance and use them for tuning k

```python
from sklearn.neighbors import KNeighborsRegressor
np.random.seed(1)

n = 300
x = np.random.uniform(0, 2*np.pi, n)
x = np.sort(x)

"""
Add noise to y
"""

y = 2*np.sin(x) + np.random.normal(size=n)

test_x = np.random.uniform(0, 2*np.pi, n)
test_x = np.sort(test_x )

test_y = 2*np.sin(test_x) + np.random.normal(size=n)

# 1-nearest neighbor
knn = KNeighborsRegressor(n_neighbors=1, weights='uniform')
knn.fit(x.reshape(-1, 1), y)
test_pred = knn.predict(test_x.reshape(-1, 1))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_by_k(ax, i, j, x, y, test_x, test_y, k, legend=False):
    
    knn = KNeighborsRegressor(n_neighbors=k, weights='uniform')
    knn.fit(x.reshape(-1, 1), y)
    
    test_pred = knn.predict(test_x.reshape(-1, 1))
    
    ax[i, j].scatter(x, y, color = 'black', s = 5, label='Training points')
    ax[i, j].plot(x, 2*np.sin(x), color = '#1ECBE1', label = 'Actual function', linewidth=3)
    ax[i, j].plot(test_x, test_pred, color = '#E1341E', alpha=0.8, label = 'Fitted function', linewidth=2)
    
    ax[i, j].set_title('k=' + str(k))
    if legend:
        ax[i, j].legend(fontsize = 12)
```

```python
sns.set_theme( font_scale=1)

figure, axis = plt.subplots(2, 4, figsize=(24, 12), constrained_layout=False)
neighbor_number_list = [1, 5, 10, 20, 50, 100, 150, 200]
import itertools
for idx, z in enumerate(list(itertools.product(range(2), range(4)))):
    i=z[0]
    j=z[1]
    plot_by_k(axis, i, j, x, y, test_x, test_y, neighbor_number_list[idx], legend=int(idx/7))
```

##### 4.5 Interpretations

- For small number of k, we can see the model are more sensitive to the artifically added noise.
- As k increase, we can see less fluctuation of the fitted line. Thus, the model is more stable. Variance decrease but bias increase. 
- For large number of k, the model has a high bias and fails to capture the information between features and targets.

## 5, Curse of dimensionality

##### 5.1 Curse of dimensionality in knn

The "curse of dimensionality" refers to the phenomenon that occurs when working with **high-dimensional data** in machine learning algorithms. 

In the context of k-Nearest Neighbors (k-NN), the curse of dimensionality refers to the fact that as the number of dimensions in the data increases, the distance between any two points in the data becomes larger and less meaningful. **(Especially for sparse data)**. 

This can **make it difficult for the k-NN algorithm to identify the nearest neighbors** of a given data point and make accurate predictions.


##### 5.2 An Example

Imagine you have a 2D dataset with points randomly distributed in a unit square. In this case, finding the nearest neighbors of a point is relatively straightforward. However, if you increase the number of dimensions to 10, the points become randomly distributed in a unit 10-dimensional cube. The distance between any two points in this high-dimensional space becomes much larger, making it more difficult to identify their nearest neighbors.



![section5.png](section5.png)

In practical applications, the curse of dimensionality can lead to overfitting and poor performance of the k-NN algorithm. One way to mitigate the curse of dimensionality is to reduce the number of dimensions in the data through techniques such as feature selection or dimensionality reduction. Another way is to use k-NN algorithms specifically designed for high-dimensional data, such as kd-trees or ball trees.

## 5, KNN is 'lazy'

The k-Nearest Neighbors (k-NN) algorithm is a lazy learning algorithm, meaning that it doesn't build a model upfront, but instead waits to make predictions until a new sample is encountered.

This leads to the main limitations of knn: it can be **computationally expensive**, especially if the training dataset is large. This is because for each new sample, the algorithm needs to calculate the distances between the new sample and all the training samples in order to find the k nearest neighbors.
