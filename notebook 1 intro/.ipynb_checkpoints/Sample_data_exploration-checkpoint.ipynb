```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### For the detailed information of the respective columns, please check the file pulse2022_data.dictionary_CSV_51 file under the same directory

# pulse2022_puf_51.csv

```python
df_pulse2022_puf = pd.read_csv("pulse2022_puf_51.csv")
```

```python
df_pulse2022_puf.head()
```

```python
for ele in df_pulse2022_puf.columns:
    print(ele)
```

## Distribution of some selected features

```python
selected_columns = ['TBIRTH_YEAR','LONGCOVID','RRACE']
for ele in selected_columns:
    c = df_pulse2022_puf[ele]
    plt.figure(figsize=(8,6))
    plt.hist(c, weights=np.ones(len(c)) / len(c)*100,alpha=1)
    plt.xlabel("%s"%(ele), size=14)
    plt.ylabel("Percentage", size=14)
    plt.title("Frequency percentage distribution of %s"%(ele))
    plt.legend(loc='center right')
```

# Pulse2022_repwgt_puf_51.csv

## This is about the contribution weight of each sampled household

```python
df_pulse2022_repwgt_puf = pd.read_csv("pulse2022_repwgt_puf_51.csv")
```

```python
df_pulse2022_repwgt_puf.head()
```

```python
for ele in df_pulse2022_repwgt_puf.columns:
    print(ele)
```

### HWEIGHT: Household level weight,The household weight for a particular household is the inverse of its household selection probability multiplied by the inverse of the household response rate in the stratum.

### PWEIFGHT: Person level weight,The person-level weight for each sampled person in the household is the product of the. final household weight and several weight adjustment factors

```python
selected_columns = ['HWEIGHT1','PWEIGHT1']
for ele in selected_columns:
    c = df_pulse2022_repwgt_puf[ele]
    bins = np.arange(0,10000,100)
    plt.figure(figsize=(8,6))
    plt.hist(c, bins=bins,weights=np.ones(len(c)) / len(c)*100,alpha=1)
    plt.xlabel("%s"%(ele), size=14)
    plt.ylabel("Percentage", size=14)
    plt.title("Frequency percentage distribution of %s"%(ele))
    plt.legend(loc='center right')
```
